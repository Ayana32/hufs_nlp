```python
def extract_word(text):
    hangul = re.compile('[^가-힣|a-zA-Z]') 
    result = hangul.sub(' ', text) 
    return result
```

```python
print("추출 전 : ",cm1['Review'].iloc[107])
print("추출 후 : ",extract_word(cm1['Review'].iloc[107]))
```

```python
cm1['Review'] = cm1['Review'].apply(lambda x:extract_word(x))
```

```python
okt = Okt()
words = " ".join(review_data.tolist())
words = okt.morphs(words,stem=True)

print(words)
```

```python
remove_two_word = [x for x in words if len(x)>2]
len(remove_two_word)
```

```python
with open('stopwords.txt', 'r') as f:
    list_file = f.readlines()
stopwords = list_file[0].split(",")
remove_stopwords = [x for x in remove_two_word if x not in stopwords]
len(remove_stopwords)
```

```python
compile = re.compile("\W+")
for i in range(len(remove_stopwords)):

    a = compile.sub(" ",remove_stopwords[i])
    remove_stopwords[i] = a.upper()
print(remove_stopwords)
```

```python
top10_words = pd.Series(remove_stopwords).value_counts().head(10)
print("날씨의아이")
print(top10_words)
```

```python
for item, count in Counter(remove_stopwords).most_common():
    if count==1:
        print(item)
```

```python
minimum_count = 3
more_than_three_time= []
for i in tqdm(range(len(remove_stopwords))):
    tmp = remove_stopwords[i]
    if remove_stopwords.count(tmp) >= minimum_count:
        more_than_three_time.append(tmp)
```

```python
def preprocess(text, word_list):
    okt = Okt()
    text = okt.morphs(text, stem = True)
    text = [x for x in text if x in word_list]
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(tokenizer = lambda x: preprocess(x,word_list))
data_features = vectorizer.fit_transform(cm1['Review'].tolist())
list_of_word = vectorizer.get_feature_names()
```

```python
count_list = data_features.toarray().sum(axis=0)
print(count_list)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_vectorizer = TfidfTransformer()
tf_idf_vect = tfidf_vectorizer.fit_transform(data_features)
```

```python
print(tf_idf_vect)
```

```python
vocab = vectorizer.vocabulary_
```

```python
# 단어들의 사전 인덱스를 이용하여 원래 단어를 검색하는 매핑 딕셔너리
index_to_word = { v:k for k, v in vocab.items() } 
```

```python
invert_index_vectorizer = {v: k for k, v in vectorizer.vocabulary_.items()}
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# LDA 모델링 객체를 생성 (토픽 갯수를 2로 설정: 긍정/부정)
lda = LatentDirichletAllocation(n_components=2)  

# TF-IDF 벡터를 입력하여 모델 학습 
lda.fit(tf_idf_vect)
```

```python
# 토픽 모델링 결과를 담고 있는 배열의 형태 : (2개의 토픽, 2157개의 단어)
print(lda.components_.shape)
```

```python
# 2157개의 단어 중에서, 토픽 별로 가장 중요도가 높은 단어를 10개씩 출력

for idx, topic in enumerate(lda.components_):
    print(f"토픽 유형 {idx+1}:", [(index_to_word[i], topic[i].round(3)) for i in topic.argsort()[:-11:-1]])
```
